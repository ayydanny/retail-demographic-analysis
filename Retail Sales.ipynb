```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("retail_sales_dataset.csv")
```

```python
conn = sqlite3.connect("retail.db")
cursor = conn.cursor()
df.to_sql("retail", conn, if_exists="replace", index=False)
```

```python
demographic_revenue = pd.read_sql("""
SELECT
CASE
WHEN age >= 18 AND age <= 24 THEN "18-24"
WHEN age >= 25 AND age <= 34 THEN "25-34"
WHEN age >= 35 AND age <= 44 THEN "35-44"
WHEN age >= 45 AND age <= 54 THEN "45-54"
ELSE "55-64"
END AS "Age Category",
Gender,
"Product Category",
SUM("Total Amount") AS "Total Amount",
SUM(Quantity) AS Quantity,
AVG("Price per Unit") AS "Avg Price per Unit"
FROM retail
GROUP BY "Age Category", Gender, "Product Category"
""", conn)

demographic_revenue.to_csv("demographic_revenue.csv", index=False)
```
